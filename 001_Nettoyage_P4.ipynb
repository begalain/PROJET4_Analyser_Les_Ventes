```python
test = transactions['date'].tolist()
tra10 = transactions.copy()
tra10['test'] = [True if 'test' in a else False for a in test]
tra11 = tra10[tra10['test'] == True]
tra11 = tra11.groupby(['test', 'session_id', 'client_id', 'id_prod']).count()
tra11
```

<p style='color:#88cc8d;text-align:center'>
    Remarque: Les 200 lignes du produit suspect T_0 correspondent aux 200 lignes de la session suspecte s_0, qui correspondent aux 200 lignes des clients supects ct_0 et ct_1, qui correspondent aux 200 lignes de dates tests.<br/>
    Il convient de retirer ces 200 lignes qui sont des tests.
</p>

<h3 style='color:#8677b2'>Retrait des 200 lignes tests</h3>

```python
transactions['test'] = [True if 'test' in a else False for a in test]
clean_transactions = transactions[transactions['test'] == False].copy()
clean_transactions = clean_transactions.drop(columns='test')
```

<h3 style='color:#8677b2'>Création des colonnes jour_sem, nom_mois, heure et partie_jour<br/>
Date devient session_date<br/>
La colonne date devient la date au format YYYY MM DD</h3>

```python
clean_transactions['session_date'] = pd.to_datetime(clean_transactions['date'])

clean_transactions['jour_sem'] = clean_transactions['session_date'].map(lambda d: d.day)
clean_transactions['nom_mois'] = clean_transactions['session_date'].map(lambda d: d.month)
clean_transactions['heure'] = clean_transactions['session_date'].map(lambda d: d.hour)

def partie_jour(value):
    if 14 <= value < 22:
        return 'Afternoon'
    elif 6 <= value < 14:
        return 'Morning'
    else:
        return 'Nigth'

clean_transactions['partie_jour'] = clean_transactions['heure'].map(partie_jour)

clean_transactions['date'] = clean_transactions['session_date'].dt.strftime('%Y %m %d')
clean_transactions.head()
```

```python
clean_transactions.to_csv('clean_transactions.csv', index=False)
clean_transactions.shape
```

<h3 style='color:#8677b2'>Création du dataframe des ventes sans le produit non référencé et sans les produits non vendus</h3>

<a id='deux_ajout'></a>
<h3 style='color:#8677b2'>Ajout de la colonne taille panier</h3>

[Haut de page](#sommaire)

```python
panier = vente.groupby(['session_id'], as_index=False).count()

panier = panier[['session_id', 'id_prod']]

panier.rename(columns={'session_id':'session_id', 'id_prod':'taille_panier_session'}, inplace=True)

vente1 = vente.merge(panier, on='session_id')
vente1.head()
```

<h3 style='color:#8677b2'>Ajout des colonne achat_mensuel, achat_annuel et freq_mens</h3>

```python
vente2 = vente.groupby(['client_id', 'nom_mois']).count()

achat_mensuel = vente2[['session_id']]

achat_annuel = vente['client_id'].value_counts()

achat = achat_mensuel.join(achat_annuel, on='client_id')

achat = achat.rename({'client_id':'client_id', 'nom_mois':'nom_mois',
                              'session_id':'achat_mens', 'client_id':'achat_an'},
                  axis='columns')

achat['freq_mens'] = achat['achat_mens'] / achat['achat_an']

achat = achat.reset_index()
vente3 = vente1.merge(achat, on=['client_id', 'nom_mois'])
vente3.head()
```

```python
vente3.to_csv('vente_totale.csv', index_label=False)
```
